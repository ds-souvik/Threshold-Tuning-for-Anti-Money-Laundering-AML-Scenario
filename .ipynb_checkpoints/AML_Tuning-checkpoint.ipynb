# 1. Create a Data Quality Report

```python
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline


pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score
from sklearn.model_selection import train_test_split, cross_val_score, KFold
import seaborn as sns

def data_quality_report(df):
    """
    Generate a comprehensive data quality report for a DataFrame.
    The report includes statistical metrics and analysis on distribution skewness and outliers.

    Parameters:
    df (pd.DataFrame): DataFrame to analyze.

    Returns:
    pd.DataFrame: Data quality report.
    """
    report = pd.DataFrame()
    for column in df.columns:
        report.at[column, 'Columns'] = column
        report.at[column, 'Missing Values'] = df[column].isna().sum()
        report.at[column, '% Missing Values'] = (df[column].isna().mean() * 100).round(2)
        report.at[column, 'Unique Values'] = df[column].nunique()
        report.at[column, 'Data Type'] = df[column].dtype

    # Compute statistics for numerical columns
    numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
    for col in numerical_cols:
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1
        report.at[col, 'Mean'] = df[col].mean()
        report.at[col, 'Median'] = df[col].median()
        report.at[col, 'Std Deviation'] = df[col].std()
        report.at[col, 'Min'] = df[col].min()
        report.at[col, 'Max'] = df[col].max()
        report.at[col, 'q1 Outliers'] = ((df[col] < (q1 - 1.5 * iqr)).sum())
        report.at[col, 'q3 Outliers'] = ((df[col] > (q3 + 1.5 * iqr)).sum())
        skewness = df[col].skew()
        if skewness > 1:
            report.at[col, 'Distribution'] = 'Right skewed'
        elif skewness < -1:
            report.at[col, 'Distribution'] = 'Left skewed'
        else:
            report.at[col, 'Distribution'] = 'Normal'

    return report

def export_report_to_excel(report, filename):
    """
    Export the data quality report to an Excel file, applying column width adjustments for readability.

    Parameters:
    report (pd.DataFrame): Data quality report.
    filename (str): Filename for the Excel file.
    """
    with pd.ExcelWriter(filename, engine='xlsxwriter') as writer:
        report.to_excel(writer, sheet_name=filename, index=False)
        worksheet = writer.sheets[filename]
        for column in report:
            col_idx = report.columns.get_loc(column)
            col_width = max(report[column].astype(str).map(len).max(), len(column)) + 1
            worksheet.set_column(col_idx, col_idx, col_width)

    print(f"Data quality report exported successfully to {filename}")

def main():
    """
    Load the dataset, generate a data quality report with advanced metrics, and export it to Excel.
    """
    # Load data
    acct_data = pd.read_csv('data/accounts.csv', usecols=['ACCOUNT_ID', 'IS_FRAUD'])
    txn_data = pd.read_csv('data/transactions.csv', usecols=['SENDER_ACCOUNT_ID', 'RECEIVER_ACCOUNT_ID', 'TX_AMOUNT'])
    #df = pd.read_csv("data/accounts.csv")  # Ensure this is correctly set to your dataset path
    report = data_quality_report(acct_data)
    export_report_to_excel(report, "account_data.xls")
    
    #df = pd.read_csv("data/transactions.csv")  # Ensure this is correctly set to your dataset path
    report = data_quality_report(txn_data)
    export_report_to_excel(report, "transactions_data.xls")

if __name__ == "__main__":
    main()
```

# 2. Generate a threshold_summary_file that consists of percentile distributions, thresholds and evaluation metrics against each thresholds

```python
acct_data = pd.read_csv('data/accounts.csv', usecols=['ACCOUNT_ID', 'IS_FRAUD'])
txn_data = pd.read_csv('data/transactions.csv', usecols=['SENDER_ACCOUNT_ID', 'RECEIVER_ACCOUNT_ID', 'TX_AMOUNT'])

# Data Preprocessing
acct_data.rename(columns={'IS_FRAUD': 'SUSPICIOUS'}, inplace=True)
acct_data['SUSPICIOUS'] = acct_data['SUSPICIOUS'].astype(int)

# Aggregate transaction data by account
credit_summary = txn_data.groupby('RECEIVER_ACCOUNT_ID', as_index=False).agg({'TX_AMOUNT': ['sum', 'count'], 'SENDER_ACCOUNT_ID': 'nunique'})
credit_summary.columns = ['ACCOUNT_ID', 'DEPOSIT_AMT', 'DEPOSIT_COUNT', 'DEPOSIT_CTPY']

debit_summary = txn_data.groupby('SENDER_ACCOUNT_ID', as_index=False).agg({'TX_AMOUNT': ['sum', 'count'], 'RECEIVER_ACCOUNT_ID': 'nunique'})
debit_summary.columns = ['ACCOUNT_ID', 'DEBIT_AMT', 'DEBIT_COUNT', 'DEBIT_CTPY']

acct_summary = acct_data.merge(credit_summary, on='ACCOUNT_ID', how='left').merge(debit_summary, on='ACCOUNT_ID', how='left').fillna(0).drop(columns=['ACCOUNT_ID'])
acct_summary.head()

# Split the data into training and testing sets
acct_summary_train = pd.concat([
    acct_summary[acct_summary['SUSPICIOUS'] == 0].sample(frac=0.6, random_state=721),
    acct_summary[acct_summary['SUSPICIOUS'] == 1].sample(frac=0.6, random_state=721)]).sort_index()

acct_summary_test = acct_summary[~acct_summary.index.isin(acct_summary_train.index)]

print('Train Set\n', acct_summary_train['SUSPICIOUS'].value_counts(), '\n')
print('Test Set\n', acct_summary_test['SUSPICIOUS'].value_counts())

# Threshold Tuning
def tune_thresholds(df, target_feature):
    """
    Tune thresholds for the target feature to identify optimal cut-offs for alerts.

    Parameters:
    df (pd.DataFrame): The DataFrame containing training data.
    target_feature (str): The feature for which to tune thresholds.

    Returns:
    pd.DataFrame: A DataFrame containing tuning results for each percentile.
    """
    summary_df = pd.DataFrame()
    
    for i in range(1, 100):
        q = i / 100
        threshold = df[target_feature].quantile(q=q)

        summary = {
            'feature': target_feature,
            'percentile': q,
            'threshold': threshold,
            'no_of_alerts': (df[target_feature] >= threshold).sum(),
            'tp_count': ((df[target_feature] >= threshold) & (df['SUSPICIOUS'] == 1)).sum(),
            'fp_count': ((df[target_feature] >= threshold) & (df['SUSPICIOUS'] == 0)).sum(),
            'tn_count': ((df[target_feature] < threshold) & (df['SUSPICIOUS'] == 0)).sum(),
            'fn_count': ((df[target_feature] < threshold) & (df['SUSPICIOUS'] == 1)).sum()
        }

        summary_df = pd.concat([summary_df, pd.DataFrame(summary, index=[0])], ignore_index=True)

    return summary_df

# Perform threshold tuning on training data
target_feature = 'DEPOSIT_AMT'
summary_train_df = tune_thresholds(acct_summary_train, target_feature)

# Evaluate thresholds using multiple metrics
summary_train_df['tp_rate'] = summary_train_df['tp_count'] / (summary_train_df['tp_count'] + summary_train_df['fn_count'])
summary_train_df['fp_rate'] = summary_train_df['fp_count'] / (summary_train_df['fp_count'] + summary_train_df['tn_count'])
summary_train_df['j_stat'] = summary_train_df['tp_rate'] - summary_train_df['fp_rate']
summary_train_df['precision'] = summary_train_df['tp_count'] / (summary_train_df['tp_count'] + summary_train_df['fp_count'])
summary_train_df['recall'] = summary_train_df['tp_count'] / (summary_train_df['tp_count'] + summary_train_df['fn_count'])
summary_train_df['f1'] = (2 * summary_train_df['precision'] * summary_train_df['recall']) / (summary_train_df['precision'] + summary_train_df['recall'])
export_report_to_excel(summary_train_df, "threshold_summary_file.xls")
summary_train_df.head()
```

# 3. Calculate optimum threshold using best J-statistics score and F1-score, visualize the confusion metrics and determine the best threshold to be used for the scenario

```python
# Select the best threshold based on Youden's J statistic
best_j_stat_idx = summary_train_df['j_stat'].idxmax()
best_j_stat_df = summary_train_df.loc[best_j_stat_idx]

best_threshold_by_j_stat = best_j_stat_df['threshold']
best_j_stat = best_j_stat_df['j_stat']
best_fpr_by_j_stat = best_j_stat_df['fp_rate']
best_tpr_by_j_stat = best_j_stat_df['tp_rate']

# Select the best threshold based on F1 score
best_f1_idx = summary_train_df['f1'].idxmax()
best_f1_df = summary_train_df.loc[best_f1_idx]

best_threshold_by_f1 = best_f1_df['threshold']
best_f1 = best_f1_df['f1']
best_precision_by_f1 = best_f1_df['precision']
best_recall_by_f1 = best_f1_df['recall']

# Print the best thresholds details
print("Best Threshold by Youden's J Statistic:")
print(f"Threshold: {best_threshold_by_j_stat}")
print(f"J Statistic: {best_j_stat}")
print(f"False Positive Rate: {best_fpr_by_j_stat}")
print(f"True Positive Rate: {best_tpr_by_j_stat}")

print("\nBest Threshold by F1 Score:")
print(f"Threshold: {best_threshold_by_f1}")
print(f"F1 Score: {best_f1}")
print(f"Precision: {best_precision_by_f1}")
print(f"Recall: {best_recall_by_f1}")

# Function to plot confusion matrix
def plot_confusion_matrix(y_true, y_pred, title):
    """
    Plot the confusion matrix for the given true and predicted labels.

    Parameters:
    y_true (pd.Series): The true labels.
    y_pred (pd.Series): The predicted labels.
    title (str): The title for the plot.
    """
    cm = confusion_matrix(y_true, y_pred)
    cm_df = pd.DataFrame(cm, index=['Not Suspicious', 'Suspicious'], columns=['No Alert', 'With Alert'])

    plt.figure(figsize=(10, 8))
    sns.heatmap(cm_df, cmap='Blues', fmt='g', cbar=True, annot_kws={"size": 15})
    plt.xlabel('With Alert', fontsize=15)
    plt.ylabel('Suspicious', fontsize=15)
    plt.title(title, fontsize=18)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)

    # Add text annotations
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            plt.text(j + 0.5, i + 0.5, f'{cm[i, j]}', horizontalalignment='center', 
                     color='white' if cm[i, j] > cm.max() / 2 else 'black')

    plt.show()

# Predict using the best threshold on the training data for both metrics
y_train_pred_j_stat = (acct_summary_train[target_feature] >= best_threshold_by_j_stat).astype(int)
y_train_pred_f1 = (acct_summary_train[target_feature] >= best_threshold_by_f1).astype(int)

# Plot confusion matrices for the training data
plot_confusion_matrix(acct_summary_train['SUSPICIOUS'], y_train_pred_j_stat, 
                      f'Confusion Matrix at {best_threshold_by_j_stat:.0f} Threshold (Training Data - J Stat)')

plot_confusion_matrix(acct_summary_train['SUSPICIOUS'], y_train_pred_f1, 
                      f'Confusion Matrix at {best_threshold_by_f1:.0f} Threshold (Training Data - F1 Score)')

# Evaluate the best thresholds on the test data
y_test_pred_j_stat = (acct_summary_test[target_feature] >= best_threshold_by_j_stat).astype(int)
y_test_pred_f1 = (acct_summary_test[target_feature] >= best_threshold_by_f1).astype(int)

# Calculate performance metrics for the test data for both metrics
def calculate_metrics(y_true, y_pred):
    tp = ((y_pred == 1) & (y_true == 1)).sum()
    fp = ((y_pred == 1) & (y_true == 0)).sum()
    tn = ((y_pred == 0) & (y_true == 0)).sum()
    fn = ((y_pred == 0) & (y_true == 1)).sum()

    tp_rate = tp / (tp + fn)
    fp_rate = fp / (fp + tn)
    precision = tp / (tp + fp)
    recall = tp / (tp + fn)
    f1 = (2 * precision * recall) / (precision + recall)
    j_stat = tp_rate - fp_rate

    return {
        'tp_rate': tp_rate,
        'fp_rate': fp_rate,
        'precision': precision,
        'recall': recall,
        'f1': f1,
        'j_stat': j_stat
    }

metrics_j_stat = calculate_metrics(acct_summary_test['SUSPICIOUS'], y_test_pred_j_stat)
metrics_f1 = calculate_metrics(acct_summary_test['SUSPICIOUS'], y_test_pred_f1)

# Print the test metrics for both thresholds
print("Test Metrics at Threshold by Youden's J Statistic:")
print(metrics_j_stat)

print("\nTest Metrics at Threshold by F1 Score:")
print(metrics_f1)

# Plot confusion matrices for the test data
plot_confusion_matrix(acct_summary_test['SUSPICIOUS'], y_test_pred_j_stat, 
                      f'Confusion Matrix at {best_threshold_by_j_stat:.0f} Threshold (Test Data - J Stat)')

plot_confusion_matrix(acct_summary_test['SUSPICIOUS'], y_test_pred_f1, 
                      f'Confusion Matrix at {best_threshold_by_f1:.0f} Threshold (Test Data - F1 Score)')
```

## 4. Optimizing the current methodology
### Optimization Steps
1. Review Performance Metrics: Compare recall values of the current thresholds.
2. Adjust Threshold: Choose a threshold that maximizes recall.
3. Validate: Validate the chosen threshold on test data.

```python
# Find the best threshold by maximizing recall on the training data
best_recall_idx_train = summary_train_df['recall'].idxmax()
best_recall_df_train = summary_train_df.loc[best_recall_idx_train]

best_threshold_by_recall_train = best_recall_df_train['threshold']
best_recall_train = best_recall_df_train['recall']
best_precision_by_recall_train = best_recall_df_train['precision']
best_f1_by_recall_train = best_recall_df_train['f1']

print("Best Threshold by Recall (Training Data):")
print(f"Threshold: {best_threshold_by_recall_train}")
print(f"Recall: {best_recall_train}")
print(f"Precision: {best_precision_by_recall_train}")
print(f"F1 Score: {best_f1_by_recall_train}")
```

```python
# Predict using the best threshold by recall on the training data
y_train_pred_recall = (acct_summary_train[target_feature] >= best_threshold_by_recall_train).astype(int)

# Calculate performance metrics for the training data
metrics_recall_train = calculate_metrics(acct_summary_train['SUSPICIOUS'], y_train_pred_recall)
print("\nTraining Data Metrics at Threshold by Recall:")
print(metrics_recall_train)

# Plot confusion matrix for the training data
plot_confusion_matrix(acct_summary_train['SUSPICIOUS'], y_train_pred_recall, 
                      f'Confusion Matrix at {best_threshold_by_recall_train:.0f} Threshold (Training Data - Recall)')
```

```python
# Predict using the best threshold by recall on the test data
y_test_pred_recall = (acct_summary_test[target_feature] >= best_threshold_by_recall_train).astype(int)

# Calculate performance metrics for the test data
metrics_recall_test = calculate_metrics(acct_summary_test['SUSPICIOUS'], y_test_pred_recall)
print("\nTest Data Metrics at Threshold by Recall:")
print(metrics_recall_test)

# Plot confusion matrix for the test data
plot_confusion_matrix(acct_summary_test['SUSPICIOUS'], y_test_pred_recall, 
                      f'Confusion Matrix at {best_threshold_by_recall_train:.0f} Threshold (Test Data - Recall)')
```

```python
def create_comparison_dataframe(threshold_type, threshold, train_metrics, test_metrics):
    """
    Create a DataFrame to compare metrics for different thresholds.

    Parameters:
    threshold_type (str): Type of threshold (e.g., 'J Statistic', 'F1 Score', 'Recall').
    threshold (float): The threshold value.
    train_metrics (dict): Performance metrics for the training data.
    test_metrics (dict): Performance metrics for the test data.

    Returns:
    pd.DataFrame: DataFrame containing the comparison of metrics.
    """
    metrics_df = pd.DataFrame({
        'Metric': ['Threshold', 'True Positive Rate (Recall)', 'False Positive Rate', 'Precision', 'F1 Score', "Youden's J Statistic", 'True Positives (TP)', 'False Positives (FP)', 'True Negatives (TN)', 'False Negatives (FN)'],
        f'Training Data ({threshold_type})': [
            threshold, 
            train_metrics['tp_rate'], 
            train_metrics['fp_rate'], 
            train_metrics['precision'], 
            train_metrics['f1'], 
            train_metrics['j_stat'], 
            train_metrics['tp'], 
            train_metrics['fp'], 
            train_metrics['tn'], 
            train_metrics['fn']
        ],
        f'Test Data ({threshold_type})': [
            threshold, 
            test_metrics['tp_rate'], 
            test_metrics['fp_rate'], 
            test_metrics['precision'], 
            test_metrics['f1'], 
            test_metrics['j_stat'], 
            test_metrics['tp'], 
            test_metrics['fp'], 
            test_metrics['tn'], 
            test_metrics['fn']
        ]
    })
    return metrics_df

# Updated calculate_metrics function to return absolute values as well
def calculate_metrics(y_true, y_pred):
    tp = ((y_pred == 1) & (y_true == 1)).sum()
    fp = ((y_pred == 1) & (y_true == 0)).sum()
    tn = ((y_pred == 0) & (y_true == 0)).sum()
    fn = ((y_pred == 0) & (y_true == 1)).sum()

    tp_rate = tp / (tp + fn)
    fp_rate = fp / (fp + tn)
    precision = tp / (tp + fp)
    recall = tp / (tp + fn)
    f1 = (2 * precision * recall) / (precision + recall)
    j_stat = tp_rate - fp_rate

    return {
        'tp_rate': tp_rate,
        'fp_rate': fp_rate,
        'precision': precision,
        'recall': recall,
        'f1': f1,
        'j_stat': j_stat,
        'tp': tp,
        'fp': fp,
        'tn': tn,
        'fn': fn
    }

# Validate and visualize metrics for J Statistic threshold
y_train_pred_j_stat = (acct_summary_train[target_feature] >= best_threshold_by_j_stat).astype(int)
train_metrics_j_stat = calculate_metrics(acct_summary_train['SUSPICIOUS'], y_train_pred_j_stat)
y_test_pred_j_stat = (acct_summary_test[target_feature] >= best_threshold_by_j_stat).astype(int)
test_metrics_j_stat = calculate_metrics(acct_summary_test['SUSPICIOUS'], y_test_pred_j_stat)
j_stat_comparison = create_comparison_dataframe("J Statistic", best_threshold_by_j_stat, train_metrics_j_stat, test_metrics_j_stat)

# Validate and visualize metrics for F1 Score threshold
y_train_pred_f1 = (acct_summary_train[target_feature] >= best_threshold_by_f1).astype(int)
train_metrics_f1 = calculate_metrics(acct_summary_train['SUSPICIOUS'], y_train_pred_f1)
y_test_pred_f1 = (acct_summary_test[target_feature] >= best_threshold_by_f1).astype(int)
test_metrics_f1 = calculate_metrics(acct_summary_test['SUSPICIOUS'], y_test_pred_f1)
f1_comparison = create_comparison_dataframe("F1 Score", best_threshold_by_f1, train_metrics_f1, test_metrics_f1)

# Validate and visualize metrics for Recall threshold
y_train_pred_recall = (acct_summary_train[target_feature] >= best_threshold_by_recall_train).astype(int)
train_metrics_recall = calculate_metrics(acct_summary_train['SUSPICIOUS'], y_train_pred_recall)
y_test_pred_recall = (acct_summary_test[target_feature] >= best_threshold_by_recall_train).astype(int)
test_metrics_recall = calculate_metrics(acct_summary_test['SUSPICIOUS'], y_test_pred_recall)
recall_comparison = create_comparison_dataframe("Recall", best_threshold_by_recall_train, train_metrics_recall, test_metrics_recall)

# Display the comparison DataFrames
print("\nComparison of Metrics for Different Thresholds:\n")

print("J Statistic Threshold:")
print(j_stat_comparison.to_string(index=False))

print("\nF1 Score Threshold:")
print(f1_comparison.to_string(index=False))

print("\nRecall Threshold:")
print(recall_comparison.to_string(index=False))
```
